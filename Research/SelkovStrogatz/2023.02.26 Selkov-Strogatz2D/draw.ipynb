```python
import numpy as np
import os
from tqdm import tqdm
import pyarrow as pa
import pylab as plt
import pyarrow.parquet as pq
from glob import glob
import seaborn as sns
```

```python
FOLDER = '/media/alexander/BigDisk/data/selk_strogg_2D_2023_02_26_long/'
```

```python
datas = glob(f'{FOLDER}/*/*.parquet')
len(datas)
```

```python
def draw(path,step=1000,axes = None):
    data = pq.read_table(path)['u'].to_numpy().reshape((-1,201,201))
    print(data.shape)
    for i, s in tqdm(enumerate(data[::step,:,:]), total=int(data.shape[0]/step)):
        ax = axes[i] if axes is not None else plt.gca()
        sns.heatmap(s, ax=ax)
    plt.show()
        
```

```python
fig, axes = plt.subplots(32,3, figsize=(4,44))
draw(datas[101],5, axes.flatten())
```
