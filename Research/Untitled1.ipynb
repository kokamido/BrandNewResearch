```python
import numpy as np
import pylab as plt
```

```python
new_resolver = np.arange(0.8,0.99, 0.01)
```

```python
plt.figure(figsize=(8,6))
plt.plot(new_resolver, [(1 - x ** 10) for x in new_resolver], lw=3, c='b', label='только новый')
plt.plot(new_resolver, [(1 - (0.666666666*x + 0.333333333*0.8) ** 10) for x in new_resolver], lw=3, c='r', label='новый и старый')
plt.grid()
plt.legend()
```
